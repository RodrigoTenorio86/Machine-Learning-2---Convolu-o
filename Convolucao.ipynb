```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np
from scipy import misc
```

## Teste de Filtro para Edif√≠cio

```
img = mpimg.imread('/content/veiculo_vermelho.png')
print(img)
```

```
imgplot = plt.imshow(img)
```

```
plt.grid(False)
plt.gray()
plt.axis('off')
plt.imshow(img)
plt.show()
```

```
i_transformed = np.copy(img)
size_x = i_transformed.shape[0]
size_y = i_transformed.shape[1]
```

## Filtro A

```
# Este filtro detecta bem as arestas
# Cria uma convolu√ß√£o que s√≥ passa por arestas vivas e linhas retas

filter = [ [0,1,0], [0,1,0], [1, 0, 1]]
weight  = 1
```

```
for x in range(1,size_x-1):
  for y in range(1,size_y-1):
      convolution = 0.0
      convolution = convolution + (img[x - 1, y-1] * filter[0][0])
      convolution = convolution + (img[x, y-1] * filter[0][1])
      convolution = convolution + (img[x + 1, y-1] * filter[0][2])
      convolution = convolution + (img[x-1, y] * filter[1][0])
      convolution = convolution + (img[x, y] * filter[1][1])
      convolution = convolution + (img[x+1, y] * filter[1][2])
      convolution = convolution + (img[x-1, y+1] * filter[2][0])
      convolution = convolution + (img[x, y+1] * filter[2][1])
      convolution = convolution + (img[x+1, y+1] * filter[2][2])
      convolution = convolution * weight
      if np.any((convolution<0)):
            convolution=0
      if np.any((convolution>255)):
            convolution=255
      i_transformed[x, y] = convolution
```

```
plt.gray()
plt.grid(False)
plt.imshow(i_transformed)
#plt.axis('off')
plt.show()   
```
